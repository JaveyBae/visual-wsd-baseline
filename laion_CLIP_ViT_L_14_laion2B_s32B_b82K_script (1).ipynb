## ä½¿ç”¨å›¾åƒ-å›¾åƒç›¸ä¼¼åº¦è¿›è¡Œè¯„ä¼°

## ğŸ¯ å›¾åƒ-å›¾åƒç›¸ä¼¼åº¦è®¡ç®—

ä½¿ç”¨ä¿®æ”¹åçš„ä»£ç ï¼Œä½ å¯ä»¥è®¡ç®—ä¸€å¼ å›¾åƒä¸å…¶ä»–å›¾åƒçš„ç›¸ä¼¼åº¦

````python
# ç¤ºä¾‹1: å•å¼ æŸ¥è¯¢å›¾åƒä¸å€™é€‰å›¾åƒçš„ç›¸ä¼¼åº¦è®¡ç®—
from vwsd import CLIP

# åŠ è½½ CLIP æ¨¡å‹
clip = CLIP('laion/CLIP-ViT-L-14-laion2B-s32B-b82K')

# ä½ çš„æŸ¥è¯¢å›¾åƒ (ä¾‹å¦‚ç”Ÿæˆçš„å›¾åƒ)
query_image = "image/my_generated_image.jpg"

# å€™é€‰å›¾åƒåˆ—è¡¨
candidate_images = [
    "dataset/image/test_images_resized/image1.jpg",
    "dataset/image/test_images_resized/image2.jpg",
    "dataset/image/test_images_resized/image3.jpg",
]

# è®¡ç®—å›¾åƒ-å›¾åƒç›¸ä¼¼åº¦
similarity = clip.get_image_similarity(query_image, candidate_images)

# è¾“å‡ºç»“æœ
print("ç›¸ä¼¼åº¦åˆ†æ•° (è¶Šé«˜è¶Šç›¸ä¼¼):")
for i, (cand, score) in enumerate(zip(candidate_images, similarity[0])):
    print(f"  {i+1}. {cand}: {score:.2f}")

# æ‰¾åˆ°æœ€ç›¸ä¼¼çš„å›¾åƒ
best_idx = similarity[0].index(max(similarity[0]))
print(f"\næœ€ç›¸ä¼¼çš„å›¾åƒ: {candidate_images[best_idx]}")
````

````python
# ç¤ºä¾‹2: æ‰¹é‡å¤„ç† V-WSD æ•°æ®é›†
import os
from vwsd import data_loader, CLIP
import pandas as pd

# åŠ è½½æ¨¡å‹å’Œæ•°æ®
clip = CLIP('laion/CLIP-ViT-L-14-laion2B-s32B-b82K')
data = data_loader('dataset')['en']

# å‡è®¾ä½ çš„ç”Ÿæˆå›¾åƒåœ¨ image/ æ–‡ä»¶å¤¹ä¸­
# å¹¶ä¸”å‘½åä¸º generated_0.jpg, generated_1.jpg, ...
generated_image_dir = "image"

results = []
for n, d in enumerate(data[:5]):  # å…ˆæµ‹è¯•å‰5ä¸ªæ ·æœ¬
    print(f"å¤„ç†æ ·æœ¬ {n+1}/5: {d['target phrase']}")
    
    # æ„é€ æŸ¥è¯¢å›¾åƒè·¯å¾„
    query_image = os.path.join(generated_image_dir, f"generated_{n}.jpg")
    
    # æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨
    if not os.path.exists(query_image):
        print(f"  è­¦å‘Š: å›¾åƒä¸å­˜åœ¨ {query_image}")
        continue
    
    # è®¡ç®—ä¸æ‰€æœ‰å€™é€‰å›¾åƒçš„ç›¸ä¼¼åº¦
    similarity = clip.get_image_similarity(query_image, d['candidate images'])
    
    # æŒ‰ç›¸ä¼¼åº¦æ’åº
    ranked = sorted(zip(similarity[0], d['candidate images']), 
                   key=lambda x: x[0], reverse=True)
    
    print(f"  æœ€ä½³åŒ¹é…: {os.path.basename(ranked[0][1])} (åˆ†æ•°: {ranked[0][0]:.2f})")
    
    results.append({
        'sample_id': n,
        'target_phrase': d['target phrase'],
        'best_match': os.path.basename(ranked[0][1]),
        'score': ranked[0][0]
    })

# æ˜¾ç¤ºç»“æœ
df = pd.DataFrame(results)
print("\nç»“æœæ±‡æ€»:")
print(df)
````
