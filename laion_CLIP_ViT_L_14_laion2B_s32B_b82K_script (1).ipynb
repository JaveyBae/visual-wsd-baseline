## 🎯 图像-图像相似度计算

使用修改后的代码，你可以计算一张图像与其他图像的相似度

```python
# 示例1: 单张查询图像与候选图像的相似度计算
from vwsd import CLIP

# 加载 CLIP 模型
clip = CLIP('laion/CLIP-ViT-L-14-laion2B-s32B-b82K')

# 你的查询图像 (例如生成的图像)
query_image = "image/my_generated_image.jpg"

# 候选图像列表
candidate_images = [
    "dataset/image/test_images_resized/image1.jpg",
    "dataset/image/test_images_resized/image2.jpg",
    "dataset/image/test_images_resized/image3.jpg",
]

# 计算图像-图像相似度
similarity = clip.get_image_similarity(query_image, candidate_images)

# 输出结果
print("相似度分数 (越高越相似):")
for i, (cand, score) in enumerate(zip(candidate_images, similarity[0])):
    print(f"  {i+1}. {cand}: {score:.2f}")

# 找到最相似的图像
best_idx = similarity[0].index(max(similarity[0]))
print(f"\n最相似的图像: {candidate_images[best_idx]}")
```

```python
# 示例2: 批量处理 V-WSD 数据集
import os
from vwsd import data_loader, CLIP
import pandas as pd

# 加载模型和数据
clip = CLIP('laion/CLIP-ViT-L-14-laion2B-s32B-b82K')
data = data_loader('dataset')['en']

# 假设你的生成图像在 image/ 文件夹中
# 并且命名为 generated_0.jpg, generated_1.jpg, ...
generated_image_dir = "image"

results = []
for n, d in enumerate(data[:5]):  # 先测试前5个样本
    print(f"处理样本 {n+1}/5: {d['target phrase']}")
    
    # 构造查询图像路径
    query_image = os.path.join(generated_image_dir, f"generated_{n}.jpg")
    
    # 检查文件是否存在
    if not os.path.exists(query_image):
        print(f"  警告: 图像不存在 {query_image}")
        continue
    
    # 计算与所有候选图像的相似度
    similarity = clip.get_image_similarity(query_image, d['candidate images'])
    
    # 按相似度排序
    ranked = sorted(zip(similarity[0], d['candidate images']), 
                   key=lambda x: x[0], reverse=True)
    
    print(f"  最佳匹配: {os.path.basename(ranked[0][1])} (分数: {ranked[0][0]:.2f})")
    
    results.append({
        'sample_id': n,
        'target_phrase': d['target phrase'],
        'best_match': os.path.basename(ranked[0][1]),
        'score': ranked[0][0]
    })

# 显示结果
df = pd.DataFrame(results)
print("\n结果汇总:")
print(df)
```

### 核心改动说明

在 `vwsd/embedding_clip.py` 中添加了两个新方法：

1. **`get_image_embedding(images, batch_size=None)`** 
   - 只计算图像的嵌入向量
   - 返回图像的特征表示

2. **`get_image_similarity(query_images, candidate_images, batch_size=None)`**
   - 计算查询图像与候选图像之间的相似度
   - 返回相似度矩阵 (query_size × candidate_size)
   - 分数范围约为 0-100，越高越相似

### 使用场景

- **场景1**: 你用 Stable Diffusion 等模型生成了图像，想找到与生成图像最相似的候选图像
- **场景2**: 你有一张参考图像，想在候选图像集中找到最相似的图像
- **场景3**: V-WSD 任务中使用图像检索而不是文本检索
